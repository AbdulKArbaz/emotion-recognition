```python
train_data_gen=ImageDataGenerator(rescale=1./255,rotation_range=30,shear_range=0.3,zoom_range=0.3,width_shift_range=0.4,height_shift_range=0.4,horizontal_flip= True,fill_mode='nearest')
```

```python
model.add(Conv2D(128,(2,2),padding='same',kernel_initializer='he_normal',input_shape=(rows,cols,1)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Conv2D(128,(2,2),padding='same',kernel_initializer='he_normal',input_shape=(rows,cols,1)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
```

```python
model.add(Conv2D(256,(2,2),padding='same',kernel_initializer='he_normal',input_shape=(rows,cols,1)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Conv2D(256,(2,2),padding='same',kernel_initializer='he_normal',input_shape=(rows,cols,1)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
```

```python
model.add(Flatten())
model.add(Dense(256,kernel_initializer='he_normal'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
model.add(Dense(512,kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Dropout(0.25))
```

```python
model.add(Dense(n,kernel_initializer='he_normal'))
model.add(Activation('softmax'))
```

```python
from keras.models import load_model
from time import sleep
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import cv2
import numpy as np
```

```python
cascade_classifier=cv2.CascadeClassifier(r'C:\Users\majaa\OneDrive\Desktop\images\haarcascade_frontalface_default.xml')
#eyes_cascade = cv2.CascadeClassifier(r'C:\Users\majaa\OneDrive\Desktop\haarcascade_eye.xml')
model=load_model(r'C:\Users\majaa\data\Emotion.h5')
class_labels=['Angry','Disgust','Fear','Happy','Neutral','Sad','Surprise']
cap=cv2.VideoCapture(0)
while True:
    # Grab a singleq frame of video
    ret, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = cascade_classifier.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,0),2)
        roi_gray = gray[y:y+h,x:x+w]
        roi_gray = cv2.resize(roi_gray,(48,48),fx=100,fy=100,interpolation=cv2.INTER_CUBIC)
        if np.sum([roi_gray])!=0:
            roi = roi_gray.astype('float')/255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi,axis=0)
            preds = model.predict(roi)[0]
            label=class_labels[preds.argmax()]
            label_position = (x,y)
            cv2.putText(frame,label,label_position,cv2.FONT_HERSHEY_SIMPLEX,2,(0,255,255),3)
        else:
            cv2.putText(frame,'No Face Found',(20,60),cv2.FONT_HERSHEY_SIMPLEX,2,(0,255,255),7)
    cv2.imshow('Face Detector',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Documentation

The convolutional layers used can be divided into three parts as shown in Fig 7:
The first part is the taking of input and making it work in the model in the field of feature extraction.
In this region we are provided with an image and this image is converted to an optimal format (mostly RGB )
and then it is made to work through the Conv2D layer which would find the pattern in the images with the help
of Loopy Pattern, the later stages would then follow the use of activation function, an activation function is
at the end deciding what is to be fired to the next neuron, in this paper we have used elu activation function,
ELU activation function also known as Exponential Linear Unit is a function which inorder to produce more 
accurate results coincides the cost to zero faster.Both the RELU activation function and ELU activation function
are very similar to each other except for the allowance of negative inputs in the function. This is used
along with Batch Normalization where Batch normalization is used to standardize each input from the mini batch
for training deep neural network. This actually leads to stabilizing of the learning process along with drastically
reducing training epochs of the model numerically which is necessary inorder  to train deep networks thus 
reducing the training time of the model. The end of this layer would be to use a Maxpooling function which would
summarize the average presence of a feature and the most activated presence of a feature. Multiple such layers 
are implemented to thoroughly improve the process of feature extraction. After this is done the next layers would
focus on flattening the obtained inputs from the above layers, Flattening is the conversion of input datawhich 
is in the form  of a matrix into a 1-D array or a feature vector for providing it as the input it to the preceding
layer. The output is flattened from the convolutional layers which would be in the form of a matrix to produce 
a single long feature vector which  is attached to the final classification model, finally forming  a fully 
connected layer. Then we finally arrive at the output part which would finally give the output of the model 
with the help of the softmax function.
